```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import pandas as pd
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']  # 使用微软雅黑字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
```

```python
df = pd.read_excel('山河四省.xlsx')
df = df.iloc[:, 2:]
X = df.drop(columns=['人口出生率（‰）'])  # 假设目标列的名称为 '目标列'
y = df['人口出生率（‰）']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2024)
train_dataset = pd.concat([X_train, y_train], axis=1)
test_dataset = pd.concat([X_test, y_test], axis=1)
```

```python
x = ['人均地区生产总值（元）', '就业人员数（万人）', '财政一般预算收入（亿元）', '财政一般预算支出（亿元）',
       '居民消费价格指数（上年=100）', '城镇居民人均可支配收入（元）', '建成区面积（平方公里）', '国家财政性教育经费（万元）',
       '年末参加生育保险人数（万人）']
y = ['人口出生率（‰）']
```

```python
dist_cols = 5
dist_rows = len(df.columns)
plt.figure(figsize=(6 * dist_cols, 5 * dist_rows))

i=1
for col in df.columns:
    ax = plt.subplot(dist_rows, dist_cols, i)
    ax = sns.kdeplot(train_dataset[col], color='Red', shade=True)
    ax = sns.kdeplot(test_dataset[col], color='Blue', shade=True)
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')
    ax = ax.legend(['train', 'test'])
    i += 1

plt.savefig('distribution.png')
```

```python
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 10)
train_corr = train_dataset.corr()
test_corr = test_dataset.corr()
train_corr
```

```python
cmap = sns.diverging_palette(250, 15, s=75, l=40, n=9, center="light", as_cmap=True)
colormap = plt.cm.RdBu
plt.figure(figsize=(8, 8)) 
sns.heatmap(train_corr,linewidths=1, vmin=-1, vmax=1,cbar_kws={"shrink": 0.8},
            square=True, cmap=colormap, linecolor='white', 
            annot=True,annot_kws={'size': 10},
            cbar=True,center=0)
plt.savefig('correlation.png')
```

```python
threshold = 0.3
high_correlation = set()
for i in range(len(abs(train_corr).columns)):
    for j in range(i):
        if abs(abs(train_corr).iloc[i, j]) > threshold:
            colname = abs(train_corr).columns[i]
            high_correlation.add(colname)
for i in range(len(abs(test_corr).columns)):
    for j in range(i):
        if abs(abs(test_corr).iloc[i, j]) > threshold:
            colname = abs(test_corr).columns[i]
            high_correlation.add(colname)

# 删除高相关性的特征
train_dataset.drop(high_correlation, axis=1, inplace=True)
test_dataset.drop(high_correlation, axis=1, inplace=True)
```

```python
cmap = sns.diverging_palette(250, 15, s=75, l=40, n=9, center="light", as_cmap=True)
colormap = plt.cm.RdBu
plt.figure(figsize=(8, 5)) 
sns.heatmap(abs(train_dataset.corr()),linewidths=1,vmax=1,vmin=0, 
            square=True, cmap=colormap, linecolor='white', 
            annot=True,annot_kws={'size': 10},
            cbar=True,center=0)
```
